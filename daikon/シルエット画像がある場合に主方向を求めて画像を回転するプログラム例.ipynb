# シルエット画像がある場合に主方向を求めて画像を回転するプログラム例

```python
import cv2

import matplotlib
# %matplotlib inline
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
from thin.thin import getSkelline
from radishlib import *
from copy import copy

# 処理対象画像
imgfile = 'pcis/daikonkabu.png'  # 元のカラー画像　処理には不要だが確認のため
bwimgfile = 'pics/daikonkabuBW.png' # シルエット画像（白黒、白255, 黒0）

# 表示のオンオフ
IMGON = True
```

```python
# 画像の読み込みと確認
src = cv2.imread(imgfile) 
bw = cv2.imread(bwimgfile,0)
plotimg(src,121)
plotimg(bw,122)
```

```python
# findContours は画像を上書きするのでコピーを作成して作業する
bwtmp = copy(bw)
srctmp = copy(src)
img, contours, hierarchy = cv2.findContours(bwtmp, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
areas = [cv2.contourArea(con) for con in contours ] # 領域の面積のリスト
num = np.argmax(areas) # 最大面積の領域番号。複数の領域がある場合、面積最大の領域を対象領域と考える
# 輪郭を描画する
cv2.drawContours(srctmp, contours, num, (0, 0, 255), 2, 8, hierarchy, 0)
# 輪郭データを浮動小数点型の配列に格納
X = np.array(contours[num], dtype=np.float).reshape((contours[num].shape[0], contours[num].shape[2]))
# 主成分分析を行う
mean, eigenvectors = cv2.PCACompute(X, mean=np.array([], dtype=np.float), maxComponents=1)
# 主成分方向のベクトルを描画
center = (mean[0][0], mean[0][1])
print(center)
vec = (eigenvectors[0][0], eigenvectors[0][1])
print(vec)
drawAxis(srctmp, center, vec, (255, 0, 0), 300)
plotimg(srctmp)
```

```python
deg = math.atan2(vec[1],vec[0])*180/np.pi + 180
need = int(np.sqrt(src.shape[0]**2+src.shape[1]**2))
xoff = int((need-src.shape[1])/2)
yoff = int((need-src.shape[0])/2)
tmpimg = np.zeros((need,need,3),dtype=np.uint8)
tmpimg[yoff:yoff+src.shape[0],xoff:xoff+src.shape[1]]=srctmp

mat = cv2.getRotationMatrix2D((center[0]+xoff,center[1]+yoff), deg, 1.0)
tmpimg = cv2.warpAffine(tmpimg, mat, (0,0))
plotimg(tmpimg)
```
