## 2 含まれる関数

### 描画用関数
- plotimg  画像をレイアウト指定に従って表示
- paraimage ２枚の画像を並べて表示
- drawRadishContour 輪郭線を描く
- drawAxis 画像にベクトルを描画する

### 画像処理のための関数
- cutmargin 余分なマージンをカットして画像サイズを小さくする
- dthreshold  2階調化したのち小ノイズを除去
- removeSmallArea ラベル画像から小領域を除去する
- getBigWhite 最大面積の白領域のみを画像として取り出す
- countWhitePixels 最大面積の白領域の画像と面積を返す
- getRadish ダイコンと指標の映った画像を分離し、単位長さあたりの画素数とダイコンの面積を出力
- labeling 画像を連結成分に分解し，ラベル数，ラベル画像，面積の3つの値を返す
- colorlabeling 画像の黒または白領域を連結成分ごとに色分けする
- skellist 画像から白画素の座標と距離をセットにしたデータリストを生成する
- skel2img 座標と距離のリストから画像を再構成する
- skellabel 細線化画像を分岐点でセグメントに分割し，セグメントごとにラベルづけした画像にして返す
- oneSkelton 細線化画像からヒゲを取り除いてヒゲのない画像を得る
- recalcDistanceP 基準点からの直線距離の順に並んだ点列から点列に沿って測った距離の列を作る
- radiusfunc 中心軸に測った距離を横軸、太さを縦軸としたグラフを描く
- makeNormalizedImage 曲がったダイコンを直線化した場合の画像をつくる
- getContour 最大径部より先だけ、その片側だけの輪郭線をえる
- getTipData 最大径部の先の片側だけの点列を得る
- shiftX 点列データ全体を指定したｙの値の位置が x = ０となるようにシフトする
- sizeNormalize 指定した長さを１としてデータをリサイズする
- drawAxis 画像に矢印を描く
- traceRidges1 指定点から指定された方向を最初の向きとして、なるべく鞍点を辿りながら反対側までトレース 
- traceRidges2 指定点から指定点までより内部の点を辿ってたどり着く
- curvature 点列の曲率を求める
- shiftX 　y 値が指定した値となる点のｘ座標が０となるようにデータ列を平行移動させる
- sizeNormalize 指定した画素数を１としてデータを正規化する
